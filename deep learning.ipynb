```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X = df.drop("Outcome", axis=1)
y = df[["Outcome"]]
```

```python
model = Sequential()
model.add(Dense(8, activation="relu")) # adding 8 neurons first--input layer and activation(decide center) Dense--connect all neurons
model.add(Dense(16, activation="relu")) # 2. layer and 16 neurons
model.add(Dense(32, activation="relu")) # 3. layer and 32 neurons
model.add(Dense(16, activation="relu"))
model.add(Dense(8, activation="relu"))
model.add(Dense(1, activation="sigmoid"))  # output layer sigmoid---due to binary
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"]) #adam gradient descent, 
```

```python
model.fit(X, y, epochs=250, batch_size=32, validation_split=.20, verbose=1) 
#epochs---repeat count, batch_size=768/32=24 part by part learning, verbose shows all steps
```

# DL---Regression

```python
df = pd.read_pickle("kc_house.pkl")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
X = df.drop("price", axis=1)
y = df[["price"]]
```

```python
X.head()
```

```python
y.head()
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size=.20, random_state=42)
```

```python
model = Sequential()
model.add(Dense(120, activation="relu")) # adding 8 neurons first--input layer and activation(decide center) Dense--connect all neurons
model.add(Dense(80, activation="relu")) # 2. layer and 16 neurons
model.add(Dense(64, activation="relu")) # 3. layer and 32 neurons
model.add(Dense(16, activation="relu"))
model.add(Dense(8, activation="relu"))
model.add(Dense(1)) # output layer sigmoid---due to binary
model.compile(loss="mse", optimizer="adam") #adam gradient descent, 
```

```python

history = model.fit(x_train, y_train, validation_data=(x_test,y_test), batch_size=128, epochs=1500)
```

```python
prediction = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
r2_score(y_test, predict)
```

```python
mean_squared_error(y_test, predict)**.5
```

```python
plt.plot(history.history["loss"], label="LOSS")
plt.plot(history.history["val_loss"], label="VALIDATION LOSS")
plt.legend();
```
